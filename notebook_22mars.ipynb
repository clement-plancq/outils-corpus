```python
import requests, json
url = "http://apps.lattice.cnrs.fr/readab/json/"
param = dict(url='https://www.nytimes.com/2021/03/21/obituaries/nawal-el-saadawi-dead.html')
resp = requests.get(url=url, params=param)
data = json.loads(resp.content)
text = data["text"]
```

```python
data["title"]
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
from collections import Counter

counter = Counter()
doc = nlp(text)
for tok in doc:
    if tok.pos_ in ['PUNCT', 'ADP', 'AUX', 'DET', 'SPACE', 'PRON', 'CCONJ', 'PART']:
        continue
    counter[tok.text] += 1

counter.most_common()
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
